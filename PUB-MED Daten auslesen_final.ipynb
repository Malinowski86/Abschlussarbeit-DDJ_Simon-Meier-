```python
#Wir bauen die Funktion "Gender Identification by name using NLTK" mithilfe von Testlisten 
import random 
from nltk.corpus import names 
import nltk 
  
def gender_features(word): 
    return {'last_letter':word[-1]} 
  
# preparing a list of examples and corresponding class labels
male_names = [(name, 'male') for name in names.words('male.txt')]
female_names = [(name, 'female') for name in names.words('female.txt')]

labeled_names = male_names + female_names
            
random.shuffle(labeled_names) 
  
from nltk import NaiveBayesClassifier

# Extract the features using the `gender_features()` function.
featuresets = [(gender_features(n), gender) for (n, gender) in labeled_names]

# Split the dataset into train and test set.
train_set, test_set = featuresets[500:], featuresets[:500]

# Train a Naive Bayes classifier
classifier = NaiveBayesClassifier.train(train_set)
```

```python
print(nltk.classify.accuracy(classifier, test_set))
```

```python
#Wir testen den Gender-Classifier mit verschiednenen Namen 
gender_classifier_test = classifier.classify(gender_features('Althaus, Christian'))
```

```python
#Wir holen uns alle Autoren raus
df = pd.DataFrame(df_covid19_Switzerland)
authorlist_full = df['Authorlist_full'].tolist()
```

```python
# Wir streichen die Einzelbuchstaben am Ende weg
import re
author_regex = '\w*,.\w*'
re.findall(author_regex, 'Boon, Gudula J A M')
```

```python
def get_clean_name(name):
    author_regex = '\w*,.\w*'
    try:
        return re.findall(author_regex, name)[0]
    except:
        return name
```

```python
out = []
for authors in authorlist_full:
    tmp = {}
    for author in authors:
        tmp[author] = get_clean_name(author)
    out.append(tmp)
out
```

```python
# Wir kombinieren jetzt Regex mit dem Gender Classifier
out = []
for authors in authorlist_full:
    tmp = {}
    for author in authors:
        if len(author) < 3:
            tmp[author] = "unknown"
        else:
            tmp[get_clean_name(author)] = (classifier.classify(gender_features(get_clean_name(author))))
    out.append(tmp)
```

```python
out
```

```python
### Wir müssen das irgendwie zählen können pro Studie, wir machen female und male Prozente 
male_count_out = []
female_count_out = []
unknown_count_out = []
percent_female_out = []
percent_male_out = []
percent_unknown_out = []
for authors in authorlist_full:
    male_count = 0
    female_count = 0
    unknown_count = 0
    for author in authors:
        try:
            first_name = get_clean_name(author).split(",")[1] # author.split(",")[1]
        except:
            first_name = "NA"
        if  len(first_name) < 3:
            result = "unknown"
        else:
            result = classifier.classify(gender_features(get_clean_name(author)))
        print("first name %s result %s" %(first_name,result))
        if result == "male":
            male_count += 1
        elif result == "unknown":
            unknown_count += 1
        else:
            female_count += 1
    try:
        percent_female = female_count/float(male_count+female_count+unknown_count)
    except:
        percent_female = -1
    try:
        percent_male = male_count/float(male_count+female_count+unknown_count)
    except:
        percent_male = -1
    try:
        percent_unknown = unknown_count/float(male_count+female_count+unknown_count)
    except:
        percent_unknown = -1
        
    male_count_out.append(male_count)
    female_count_out.append(female_count)
    unknown_count_out.append(unknown_count)
    percent_female_out.append(percent_female)
    percent_male_out.append(percent_male)
    percent_unknown_out.append(percent_unknown)
```

```python
df_covid19_Switzerland["Prozentsatz Autorinnen pro Paper"] = percent_female_out
df_covid19_Switzerland["Prozentsatz Autoren pro Paper"] = percent_male_out
df_covid19_Switzerland["Prozentsatz unbekannt pro Paper"] = percent_unknown_out

df_covid19_Switzerland["Anzahl Autoren pro Paper"] = male_count_out
df_covid19_Switzerland["Anzahl Autorinnen pro Paper"] = female_count_out
df_covid19_Switzerland["Anzahl unbekannt pro Paper"] = unknown_count_out
```

```python
#Wir holen uns, um die Anzahl Forschern an Schweizer Insitituen berechenen zu können, aus den Metadaten 
#alle Institute und Autoren raus
entries = []
i = 0
with open('pubmed-switzerlan-set_aktualisiert.txt') as openfile:
    next_entry = False
    entry = {"authors": "", "institutes": ""}
    for line in openfile:
        i += 1
        if i >= 10000:
            break
        #print(line)
        if "PMID-" in line:
            entry["PMID"] = line.split("-")[1]
        if "TI" in line:
            #print(line)
            try:
                entry["title"] = line.split("  - ")[1] 
            except:
                entry["title"]  = "NA"
        if "AU" in line:
            entry["authors"] += line.split("-")[1]
        if "FAU" in line:
            entry["authors"] += line.split("-")[1]
        if "AD" in line:
            try:
                entry["institutes"] += line.split("-")[1]
            except:
                entry["institutes"] += "NA"
        if line == "\n":
            entries.append(entry)
            entry = {"authors": "", "institutes": ""}
pd.DataFrame(entries)
```
